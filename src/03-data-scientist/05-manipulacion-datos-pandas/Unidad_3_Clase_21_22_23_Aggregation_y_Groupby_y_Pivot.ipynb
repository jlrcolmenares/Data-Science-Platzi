Trabajar con funciones

```
def mean_eur2usd(x):
  return np.mean(x)*1.12

mean_eur2usd(100)
```

```
# para aplicar la función al DF que tenemos en groupby
df.groupby('sex')[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```
# nosotros también podemos definir la funciones in line por medio del operador lambda
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(lambda x: np.mean(x)*1.12)
```

```
# en un diccionario vamos a escribir las funciones que se deben utilizar para hacer este proceso
dict_agg = {'tip':[min,max], 'total_bill':[np.mean, mean_eur2usd]}
dict_agg
```

```
df.groupby(['sex','time'])[['total_bill','tip']].agg(dict_agg)
# fijate como el sistema el orden en que llames a la columna. 
# la función aggregate(agg) aplica la función a las columnas correctas
```

```
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 22
# seleccionamos todas las facturas que tengan un valor mayor a 22
```

```
# dentro de la función filter se carga la condición
df.groupby(['sex','time']).filter(f_filter)
```

```
# al dataframe le agrego una columna
df['ones'] = 1
df
```

```
# vamos a aplicar una agrupación con variables categóricas
df_g = df.groupby(['sex','smoker'])[['ones']].sum()
df_g # vemos que por medio de la sum() obtenemos un conteo  TRUCAZO
```

```
df_g.groupby(level=0).apply( #vamos a pedir que lo agrupe por el valor cero
  lambda x:
    x/ x.sum()*100
) # con esta función lo que obtenemos es un porcentaje de quienes fuman y quienes no dentro de hombre y mujeres
```

```
# podemos segmentar los datos en subgrupos y dejar que pandas determine las mejores agrupaciones
pd.cut(df['total_bill'],bins=3)
# Aqui vemos que la totalbill se divide en 3 categorias
# y los rangos de las categorías se muesstran abaajo
# [(3.022, 18.983] < (18.983, 34.897] < (34.897, 50.81]]
```

```
# también podemos definir nosotros mismo los rangos por medio de una lista
pd.cut(df['total_bill'],bins=[3,18,35,60]).value_counts()
```

```
# podemos agregar una nueva columna al dataframe por medio de
df['bin_total'] = pd.cut(df['total_bill'],bins=[3,18,35,60])
df
```

```
# sobre esta categoría también podemos haccer groupby
df.groupby(['time','bin_total'])[['ones']].count()
```

# Tabla dinámicas / Pivot Table

```
# generamos un data frame
df.groupby(['sex','time'])[['total_bill']].mean()
```

```
# vamos a darle reset a los índices del archivo
df_gp = df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
```

```
# aqui vamos a declarar cual va a ser mi nuevo índice y cuales van a ser mis columnas
df_gp.pivot_table(values='total_bill',index='sex', columns='time')
```

```
# tu puedes aplicar esa misma conversión sin necesidad de definir df_gp
# ahora, si no queremos que se calcule el promedio, puedes describir otra función
df.pivot_table(values='total_bill',index='sex', columns='time', aggfunc=np.median)
```

```
# por medio de listas podemos generar el pivot para varias funciones
df_pivot = df.pivot_table(values='total_bill',index='sex', columns='time', aggfunc=[np.median,np.std])
# aqui obtengo 2 estimadores estadístivos
df_pivot
```

```
df_pivot.unstack().reset_index()
```
